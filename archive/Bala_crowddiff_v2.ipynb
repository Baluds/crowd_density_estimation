```python
import os
import matplotlib.pyplot as plt
from PIL import Image

# Specify the folder path (update with your folder name if needed)
folder_path = '/content/drive/MyDrive/cs682_project/crowddiff/out_v5'

# Get all image files from the folder
image_files = [f for f in os.listdir(folder_path) if f.endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

# Set the number of columns for the grid display
num_columns = 5
num_rows = (len(image_files) + num_columns - 1) // num_columns

# Create a figure to display images in a grid
plt.figure(figsize=(15, 15))

# Iterate through image files and add them to the plot
for idx, image_file in enumerate(image_files):
    img_path = os.path.join(folder_path, image_file)
    img = Image.open(img_path)

    plt.subplot(num_rows, num_columns, idx + 1)
    plt.imshow(img)
    plt.axis('off')  # Hide axis

plt.tight_layout()
plt.show()
```
